```python
import os
from pyngrok import ngrok, conf
import nest_asyncio
import threading
import uvicorn

# Allow any frontend to connect!
os.environ["ALLOWED_ORIGINS"] = "*"

# 1. Setup Ngrok (Replace YOUR_NGROK_TOKEN with an auth token from https://dashboard.ngrok.com)
NGROK_TOKEN = "YOUR_NGROK_TOKEN"
conf.get_default().auth_token = NGROK_TOKEN

# 2. Open tunnel
public_url = ngrok.connect(8000).public_url
print("=================================================")
print(f"🚀 YOUR API URL IS: {public_url}")
print("➡️ COPY THIS URL and paste it into your frontend!")
print("=================================================")

# 3. Run FastAPI server in the background
nest_asyncio.apply()
uvicorn.run("main:app", host="0.0.0.0", port=8000)
```
